## Named Entities with spaCY

There are Named Entity Taggers available through popular nlp libraries like "nltk" and "spacy". Each of these libraries follows a different approach to solve the problem. For this task, we use spaCY library. [spaCY](https://spacy.io/api/data-formats#named-entities)

```python
#first, we calculate many different categories, later we can decide what to include and exclude

pos_family = {
    'person' : ['PERSON'],
    'norp' : ['NORP'],
    'fac' : ['FAC'],
    'org' : ['ORG'],
    'gpe' : ['GPE'],
    'loc' : ['LOC'],
    'product' : ['PRODUCT'],
    'event' : ['EVENT'],
    'work_of_art' : ['WORK_OF_ART'],
    'law' : ['LAW'],
    'language' : ['LANGUAGE'],
    'date' : ['DATE'],
    'time' : ['TIME'],
    'percent' : ['PERCENT'],
    'money' : ['MONEY'],
}
```

```python
ner_entities = ['person_count', 'norp_count', 'fac_count', 'organization_count', 'gpe_count', 'loc_count', 'product_count', 'event_count', 'workOfart_count', 'law_count', 'language_count',
                'date_count', 'time_count', 'percent_count', 'money_count']
```

```python
#function to count named entities
def check_ner(x, flag):
    cnt = 0
    try:
        nlp_text = nlp(x)
        for tup in [(word.text, word.ent_type_) for word in nlp_text]:
            ppo = list(tup)[1]
            if ppo in pos_family[flag]:
                cnt += 1
    except:
        pass
    return cnt
```

```python
named_entities = []
temp_entity_name = ''
temp_named_entity = None
for term, tag in ner_tagged:
    if tag:
        temp_entity_name = ' '.join([temp_entity_name, term]).strip()
        temp_named_entity = (temp_entity_name, tag)
    else:
        if temp_named_entity:
            named_entities.append(temp_named_entity)
            temp_entity_name = ''
            temp_named_entity = None
```

```python
c = Counter([item[1] for item in named_entities])
c.most_common()
```
